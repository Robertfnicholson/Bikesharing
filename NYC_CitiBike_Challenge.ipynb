```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
# File to load
citibike_data_to_Load = "201908-citibike-tripdata.csv"

# Read the Citi Bike Data and store into a Pandas DataFrame
citibike_data_df = pd.read_csv(citibike_data_to_Load)
```

```python
# 2. Check the datatypes of your columns. 
citibike_data_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
# Referenced code from Julie Pyle [Baylex] and file "NYC_Citibike_Challenge.ipynb" on GitHub to correct syntax.
# citibike_data_df['tripduration'] = pd.to_datetime(citibike_data_df['tripduration'], format='%Y%m%d%H%M%S')
new_tripduration = citibike_data_df.tripduration
new_tripduration_datetime = pd.to_datetime(new_tripduration, unit='s')
# Add a new column for the converted trip duration
citibike_data_df['new_tripduration'] = new_tripduration_datetime
```

```python
# 4. Check the datatypes of your columns. 
citibike_data_df.head()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
citibike_data_df.to_csv("citibike_data_2.csv", index = False)
```
